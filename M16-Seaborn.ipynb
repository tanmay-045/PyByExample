###  Ques 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. **Scatter Plot (`scatterplot`)**:
   - **Use**: Shows the relationship between two continuous variables. Useful for identifying clusters, trends, or correlations between variables.

2. **Histogram (`histplot`)**:
   - **Use**: Displays the distribution of a single numerical variable. Helps understand the frequency and spread of data points across different ranges.

3. **Bar Plot (`barplot`)**:
   - **Use**: Visualizes categorical data with rectangular bars. Suitable for comparing values across different categories, such as comparing average values or counts.

4. **Heatmap (`heatmap`)**:
   - **Use**: Displays matrix-like data where individual values are represented as colors. Helpful for visualizing correlations in a dataset or for displaying the intensity of values across two dimensions.

5. **Box Plot (`boxplot`)**:
   - **Use**: Summarizes the distribution of numerical data through quartiles. Shows potential outliers and the variability of data points across different categories or groups.

### Ques 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
fmri['region'].value_counts()
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue='event', style='region')
plt.title('FMRI Signal Over Time by Event and Region')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Passenger Class')
plt.show()

sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hueparameter for the 'cut' column of the diamonds dataset.

```python
diamonds=sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
# kde=True: Optionally adds a kernel density estimate (KDE) curve to the histogram bars for smooth representation of the distribution.
# palette='viridis': Sets the color palette for different cuts.

sns.histplot(data=diamonds, x='price', hue='cut', kde=True, palette='viridis')
plt.title('Histogram of Diamond Prices by Cut')
plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Load the flights dataset
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# Create a pivot table with months as rows and years as columns, and passengers as values
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# # Create a heatmap

# cmap='YlGnBu': Sets the color map (colormap) to 'YlGnBu' (Yellow-Green-Blue), indicating increasing passenger numbers from yellow to blue.
# annot=True: Optionally adds annotations (numbers) to each cell of the heatmap.
# fmt='d': Specifies the format of annotations as integers ('d' stands for integer format).

sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d')


# Add title and labels
plt.title('Passenger Numbers (1000s) by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Display the plot
plt.show()
```
