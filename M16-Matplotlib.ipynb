```python
# example for pie charts

import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

#### Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.



The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to arrange and display multiple plots in a grid-like layout, making it easier to compare different datasets or different aspects of the same dataset.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig,axs = plt.subplots(2,2, figsize=(6,5))

axs[0,0].plot(x,y1,color="red")
axs[0,0].set_title("Line 1")

axs[0,1].plot(x,y2,color="Yellow")
axs[0,1].set_title("Line 2")

axs[1,0].plot(x,y3,color="Purple")
axs[1,0].set_title("Line 3")

axs[1,1].plot(x,y4,color="Green")
axs[1,1].set_title("Line 4")

plt.tight_layout()
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit, color='Black', alpha=0.8, width=0.5)
plt.xlabel('Companies')
plt.ylabel('Profit (Millions)')
plt.title('Bar Plot- C v/s p')
plt.show()

plt.barh(company,profit, color='Brown', alpha=0.5)
plt.xlabel('Profit (Millions)')
plt.ylabel('Companies')
plt.title('Bar Plot(Horizontal) - C v/s p')
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


A box plot, also known as a whisker plot, is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It provides a visual summary that can reveal insights into the spread and skewness of the data, as well as identify potential outliers.

#### Components of a Box Plot:
1. **Box**: The central box represents the interquartile range (IQR), which contains the middle 50% of the data. The bottom of the box is the first quartile (Q1), and the top is the third quartile (Q3).
2. **Median Line**: A line inside the box marks the median (Q2) of the dataset.
3. **Whiskers**: The "whiskers" extend from the box to the smallest and largest values within 1.5 times the IQR from the first and third quartiles, respectively.
4. **Outliers**: Points outside the whiskers are considered outliers and are typically plotted as individual points.

#### Why is a Box Plot Used?
1. **Summarizes Data**: Provides a concise summary of the dataset's central tendency, variability, and distribution shape.
2. **Identifies Outliers**: Highlights data points that are significantly different from the rest of the data.
3. **Compares Distributions**: Useful for comparing the distribution of data across different categories or groups.
4. **Detects Skewness**: Shows the skewness of the data by the relative positions of the median, quartiles, and whiskers. 

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# Create a box plot
data = [box1, box2]

plt.boxplot(data, vert=True, patch_artist=True, labels=['Box 1', 'Box 2'])

# Customizing the plot
plt.title('Box Plot of Two Distributions')
plt.ylabel('Values')
plt.xlabel('Box Groups')

# Show the plot
plt.show()
```
